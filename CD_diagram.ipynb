```python
import pandas as pd 
import json
from src.CD_diagram import draw_cd_diagram
```

```python
with open('evaluation_110_dataset.json') as json_file:
    data = json.load(json_file)
```

```python
#we have 110 datasets with 5 classifiers without AE and 6 classifiers with AE(including NN)
def getDataset(data):
    CD_data = pd.DataFrame(columns = ['classifier_name', 'dataset_name', 'accuracy'])
    for item in data:
        for key1 in item:
            for key2 in item[key1]:
                for key3 in item[key1][key2]:
                    if key3 == 'accuracy':
                        CD_data = CD_data.append({'classifier_name': key2, 'dataset_name': key1, 'accuracy': 
                                       item[key1][key2][key3][1]}, ignore_index = True)
    return CD_data

def ReformDataset(dataset):
    New_data = pd.DataFrame(columns = ['classifier_name', 'dataset_name', 'accuracy'])
    for index, row in dataset.iterrows():
        row['classifier_name'] = row['classifier_name'] + '_' + row['dataset_name'].split('_', maxsplit = 1)[1]
        row['dataset_name'] = row['dataset_name'].split('_', maxsplit = 1)[0]
        New_data = New_data.append({'classifier_name': row['classifier_name'], 'dataset_name': row['dataset_name'],
                            'accuracy': row['accuracy']}, ignore_index = True)
        New_data.sort_values(by = ['classifier_name'], inplace = True)
    return New_data
```

```python
#the format is not suited to draw CD_diagram
CD_data = getDataset(data)
CD_data.head(11)
```

```python
#reformat
ReadyData = ReformDataset(CD_data)
ReadyData.head(5)
```

```python
draw_cd_diagram(df_perf = ReadyData)
```
